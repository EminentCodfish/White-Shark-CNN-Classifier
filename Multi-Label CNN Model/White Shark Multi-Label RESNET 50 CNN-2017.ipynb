# Import the data

Look at the labels csv

```python
df = pd.read_csv(path+'ws_metadata.csv')
df.head()
```

Create a data bunch

```python
tfms = get_transforms()
```

The ImageList contains the file names and associated labels.  The image set was split 80/20 into training and validation set.

```python
np.random.seed(42)
src = (ImageList.from_csv(path, 'ws_metadata.csv')
       .split_by_rand_pct(0.2)
       .label_from_df(label_delim=' '))
```

Rescale images to 224 x 224 and apply data augmentation.

```python
data = (src.transform(tfms, size=224)
        .databunch().normalize(imagenet_stats))
```

# Train the model - RESNET 50

This model is the resnet50 convolutional neural network, which is a robust, medium sized model. This model is pre-trained on the imagenet dataset. This first iteration of fitting will fine-tune the outputs from the imagenet classes to our classes.

Set the model architecture

```python
arch = models.resnet50
```

Our base architecture is resnet50 again, but the metrics are a little bit differeent: we use accuracy_thresh instead of accuracy. It was determined the predicition for a given class by picking the final activation that was the biggest, but here, each activation can be 0. or 1. accuracy_thresh selects the ones that are above a certain threshold (set to sigmoid(0.2) or ~0.55) and compares them to the ground truth.

```python
acc_02 = partial(accuracy_thresh, thresh=0.2)
f_score = partial(fbeta, thresh=0.2)
learn = cnn_learner(data, arch, metrics=[acc_02, f_score])
```

We use the LR Finder to pick a good learning rate.

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

Fit the model

```python
lr = 1e-2
```

```python
learn.fit_one_cycle(6, slice(lr))
```

```python
learn.save('ws_multi_label_stage-1')
```

Let's unfreeze the last couple layers and fine-tune the convolutional layers to improve the customization.

```python
learn.unfreeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(8, slice(1e-6, 1e-5))
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(4, slice(1e-6, 1e-5))
```

```python
learn.save('stage-2')
```
